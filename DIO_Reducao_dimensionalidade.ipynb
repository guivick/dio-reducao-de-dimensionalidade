Importação das bibliotecas necessárias:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

Lendo a imagem, transformando num vetor numpy e exibindo-a:

```python
imagem = mpimg.imread('./fred.jpeg')
imagem = np.array(imagem)
plt.imshow(imagem)
```

Verificando o formato da imagem:

```python
imagem.shape
```

Convertendo a imagem para tons de cinza tilizando a expressão gray = 0.2989 * R + 0.5870 * G + 0.1140 * B:

```python
imagem_tons_de_cinza = 0.2989*imagem[:,:,0] + 0.5870*imagem[:,:,1] + 0.1140*imagem[:,:,2]

imagem_tons_de_cinza.shape
```

```python
plt.imshow(imagem_tons_de_cinza, cmap='gray')
```

```python
imagem_binaria = imagem_tons_de_cinza > 127
plt.imshow(imagem_binaria, cmap='gray')
```

Exibindo as três imagens simultaneamente para comparação:

```python
plt.subplot(1, 3, 1)
plt.axis('off')
plt.imshow(imagem)
plt.subplot(1, 3, 2)
plt.axis('off')
plt.imshow(imagem_tons_de_cinza, cmap='gray')
plt.subplot(1, 3, 3)
plt.axis('off')
plt.imshow(imagem_binaria, cmap='gray')
plt.show()
```
